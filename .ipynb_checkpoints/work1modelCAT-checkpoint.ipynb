```python
import pandas as pd
from sklearn.model_selection import train_test_split 
from catboost import CatBoostClassifier
```

```python
data = pd.read_csv('crime.csv')
```

```python
X = data[['City','Crime Description', 'Victim Age', 'Victim Gender', 'Weapon Used', 'Time_sine', 'Time_cos','Month_sine','Month_cos','DoW_sine','DoW_cos']]
X.head(3)
```

```python
Y = data['Crime Domain']
Y.head(3)
```

```python
X_train, X_test,Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
Categorical = ['City', 'Victim Gender', 'Crime Description', 'Weapon Used']
```

```python
model = CatBoostClassifier(random_state=0, loss_function='MultiClass', eval_metric='Accuracy', verbose=1)
```

```python
model.fit(X_train, Y_train, cat_features=[0, 1, 3, 4])
```

```python
y_pred = model.predict(X_test)
```
